```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print("Keys of iris_dataset: \n{}".format(iris_dataset.keys()))
```

键（Key）	    描述
'target_names'	鸢尾花的类别名称（字符串数组），如 ['setosa', 'versicolor', 'virginica']。
'feature_names'	每个特征（属性）的名称（字符串数组），如萼片长度、花瓣宽度等。
'DESCR'	        数据集的完整描述（字符串），包含来源、统计信息等。
'data'	        特征数据（二维数组，形状为 (n_samples, n_features)），即样本的特征值。
'target'	    标签数据（一维数组），表示每个样本对应的类别编号（0、1、2）。

```python
print("First five columns of data:\n{}".format(iris_dataset['data'][:5]))
```
